```python
import numpy as np
import pandas as pd 
```

```python
mydata = pd.read_csv('Laliga.csv') #reading the Data 
mydata.info()
```

```python
df0 = mydata.replace('-',0) #replacing NaN with 0
df0
```

```python
x=df0.Debut
df0['Debut']=df0['Debut'].apply(pd.to_numeric)
x = df0[(df0['Debut']>1929) & (df0['Debut']<1981)] #printing teams that Debut between 1930-1980
x
```

```python
df0['Points']= df0['Points'].apply(pd.to_numeric)
df0.nlargest(5,'Points', keep= 'first') #top 5 top scoring teams 
```

```python
df0['GoalsFor']=df0['GoalsFor'].apply(pd.to_numeric)
df0['GoalsAgainst']=df0['GoalsAgainst'].apply(pd.to_numeric)

                                              #Creating function 
def Goal_diff(df0):
    return df0.GoalsFor-df0.GoalsAgainst

df0['Goal_diff']=Goal_diff(df0)    #adding Column 
df0                                           #calling Value 
```

```python
df0.head(1)
```

```python
df0.tail(1)
```

```python
df0['GamesWon'] = df0['GamesWon'].apply(pd.to_numeric)
df0['GamesPlayed'] = df0['GamesPlayed'].apply(pd.to_numeric)
#creating function 
def Winning_Percentage(df0):
    return (df0.GamesWon/df0.GamesPlayed)*100
#joining Column 
df0['Winning_Percentage'] = Winning_Percentage(df0)

df0.nlargest(5,'Winning_Percentage',keep='first') #Calling for top 5
```

```python
df0['BestPosition'] = df0['BestPosition'].apply(pd.to_numeric)
Grouped = df0.groupby('BestPosition')     #grouping by best position 
df0.head(10)
```

```python
Grouped.groups
```

```python
for Team, group in Grouped:
    print("Best Position:" +str(Team))
    print("Toatl Points:" +str(sum(group['Points'])))
    print("-------------------------------------------------------------------------------------")
```
